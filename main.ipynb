```python
from datetime import datetime, timedelta
```

```python
timestamp_format = '%Y-%m-%dT%H:%M:%S.%fZ'
blacklist_ip = []
count = 0
last_time = None
for hit in res_dict["hits"]["hits"]:
    timestamp_str = hit["_source"]["@timestamp"]
    timestamp = datetime.strptime(timestamp_str,timestamp_format)
    local_time = timestamp+timedelta(hours=8) #Asia/Taipei
    if last_time and local_time - last_time <= timedelta(seconds=5):
        count+=1
    else:
        count=1
    last_time = local_time
    if count >= 10:
        blacklist_ip.append(hit["_source"]["src_ip"])
```

```python
blacklist_ip = list(set(blacklist_ip))
blacklist_ip
```
